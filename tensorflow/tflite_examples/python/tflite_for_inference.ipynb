## 1 Create interpreter with the tflite model

```
import numpy as np
import tensorflow as tf

# Load TFLite model and allocate tensors.
interpreter = tf.lite.Interpreter(model_path="mobilenet.tflite")
interpreter.allocate_tensors()
```

## 2 Inspect the IO info


```
# Get input and output tensors.
input_details = interpreter.get_input_details()
output_details = interpreter.get_output_details()

input_details, output_details
```

## 3 feed input

```

from PIL import Image

img = np.array(Image.open('dog.png').convert('RGB').resize((224,224)) ) 

# Test model on random input data.
input_shape = input_details[0]['shape']
img = img.reshape((1,224,224,3))
  
interpreter.set_tensor(input_details[0]['index'], img)

img.shape 
```

## 4 Run

```
import time
start = time.clock()
interpreter.invoke()
end = time.clock()
end - start
```

## 5 Get output

```

output_data = interpreter.get_tensor(output_details[0]['index'])

labels = open('labels.txt','r').readlines()
type(output_data),len(labels)

# print all the class with prob > threashhold
[(x,output_data[0][x]/255.0,labels[x]) for x in range(1,1001) if output_data[0][x] > 0.05 * 255]
```
